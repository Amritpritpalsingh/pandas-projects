```python
import pandas as pd
import seaborn as sb
import numpy as np
df = sb.load_dataset("tips")
```

```python
#Task 1 

#1: Display the first 5 rows.

print(df.head())

#2: Display the last 5 rows.

print(df.tail())

#3: Show the number of rows and columns.

print(df.info())

#4: Show all column names.

print(df.columns)

#5: Check the data types of each column.

print(df.dtypes)

#6: Show summary statistics using describe().

print(df.describe())
```

```python

# print(df)

# # 2: Show first 10 rows

# print(df.head(10))#Task 2 

# 1: rows where tip > 5.

print(df[df["tip"]>5])

# 2: Select rows where time = "Dinner".

print(df[df["time"]=="Dinner"])

# 3: Select rows where 20 < total_bill < 40.

print(df[(df["total_bill"]>20) & (df["total_bill"]<40)])
# OR 
print(df.query("20<total_bill<40"))

# 4: Select rows where sex = "Female" AND smoker = "Yes".

print(df[(df["sex"]=="Female")&(df["smoker"]=="Yes")])
```

```python
#Task 3

# 1: Calculate the average tip.

print(df["tip"].mean())

# 2: Calculate the average total bill.

print(df["total_bill"].mean())

# 3: Find the average tip for Male vs Female.

print("Males:- ",df[(df["sex"]=="Male")]["tip"].mean())
print("Female:- ",df[df["sex"]=="Female"]["tip"].mean())
# 4: Find total bill collected on each day.

print(df.groupby("day")["total_bill"].sum())

# 5: Find average tip for Lunch vs Dinner.

print("Lunch:- ",df[df["time"]=="Lunch"]["tip"].mean())
print("Dinner:- ",df[df["time"]=="Dinner"]["tip"].mean())

# 6: Find average total bill by table size.

total_bill= df.groupby("size")["total_bill"].mean()
```

```python
#task 4  

# 1: Find the row with the highest tip.

print(df.loc[df["tip"].idxmax()])
#OR
print(df.nlargest(1,"tip"))

# 2: Find the row with the largest total bill.
print(df.loc[df["total_bill"].idxmax()])
#OR
print(df.nlargest(1,"total_bill"))
# 3: Find the row with the smallest tip.
print(df.loc[df["tip"].idxmin()])
#OR
print(df.nsmallest(1,"tip"))
#4: Find the day with the highest total revenue.
revByday = df.groupby("day",observed=False)["total_bill"].sum()
print(revByday.loc[revByday.idxmax()])
```

```python
#Task 5


# 1: Create a new column showing tip percentage.

df["tipPercentage"] = df["tip"]/ df["total_bill"]

print(df)

# 2: Create a column called good_tip:

df["good_tip"] = None

print(df)

# 3: "Good" if tip percentage > 20% "Normal" otherwise

df["good_tip"] = np.where(df["tipPercentage"]>.2,"Good","Normal")

```
