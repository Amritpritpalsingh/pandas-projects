```python
# ✅ TASK 1: LOAD & EXPLORE THE DATA

# 1️⃣ Load the CSV using pd.read_csv()

df = pd.read_csv("students_marks.csv")

# 2️⃣ Print:

#     First 5 rows

print(df.head())

#     Last 5 rows

print(df.tail())

#     Shape of dataset

print(df.shape)

# 3️⃣ Print:
#     Column names

print(df.columns)

#     Data types

print(df.dtypes)

# 4️⃣ Summary statistics for numeric columns

print(df.describe())

# 5️⃣ Count missing values for each column

print(df.isnull().sum())
```

```python
# ✅ TASK 2: DATA CLEANING

# 1️⃣ Replace missing numeric values with mean

df = df.fillna(df.mean(numeric_only=True))

# 2️⃣ Replace missing categorical values with mode

cat_col = df.select_dtypes(include="object").columns
for col in cat_col:
    df[col] = df[col].fillna(df[col].mode()[0])

# 3️⃣ Remove duplicate rows

df = df.drop_duplicates()

# 4️⃣ Strip extra spaces from Name column

df.columns = df.columns.str.strip()

# 5️⃣ Make Gender → category type

df["Gender"] = df["Gender"].astype("category")

# 6️⃣ Verify dataset after cleaning

print(df)
```

```python
# ✅ TASK 3: CREATE NEW COLUMNS

# 1️⃣ Total = Math + Science + English + History

df["Total"] = df["Math"]+df["Science"]+df["English"]+df["History"]

# 2️⃣ Average = Total / 4

df["Average"] = df["Total"]/4

# 3️⃣ Create a Grade column:


def grade(marks):
    if marks >= 90: return "A"
    elif marks >= 80: return "B"
    elif marks >= 70: return "C"
    elif marks >= 60: return "D"
    else: return "F"

# A = 90+

# B = 80–89

# C = 70–79

# D = 60–69

# F = <60

df["Grade"] = df["Average"].apply(grade)

# OR

# Methord 2:

# df["Grade"] = pd.cut(
#     df["Total"],
#     bins=[0,60,70,80,90,float("inf")]
#     labels=["F","D","C","B","A"]
#     right=False
# )


# 4️⃣ Add Pass_Fail:

# Pass = Average >= 40

# Fail = otherwise

df["Pass_Fail"] = pd.cut(
    df["Average"],
    bins=[0,40,float("inf")],
    labels=["Fail","Pass"],
    right=False

)
```

```python
# ✅ TASK 4: ANALYSIS TASKS

# 1️⃣ Class average of all students

class_avg = df["Total"].mean()


# 2️⃣ Highest & lowest average scorer

lowest_avg_scr = df.iloc[df["Average"].idxmin()]

highest_avg_scr = df.iloc[df["Average"].idxmax()]

# 3️⃣ Subject-wise average marks:
#     Math avg

math_avg = df["Math"].mean()

#     Science avg

science_avg = df["Science"].mean()

#     English avg

english_avg = df["English"].mean()

#     History avg

history_avg = df["History"].mean()

# 4️⃣ Count how many students got each Grade (A, B, C, D, F)

count = df.groupby("Grade")["Student_ID"].count()


# 5️⃣ Top 3 performers

top_per = df.nlargest(3,"Total")


# 6️⃣ Gender-wise average marks

gen_w_avg = df.groupby("Gender",observed=False)["Average"].mean()

```

```python
# ✅ TASK 5: ADVANCED ANALYSIS

# 1️⃣ Which subject is the strongest (highest average)?

strg_avg = df[["Math","English","Science","History"]].mean().idxmax()


# 2️⃣ Which subject is the weakest?

wkst_avg = df[["Math","English","Science","History"]].mean().idxmin()

# 3️⃣ How many students passed vs failed?

p_f_s = df.groupby("Pass_Fail",observed=False)["Student_ID"].count()


# 4️⃣ Which gender performs better overall?

gen_per = df.groupby("Gender",observed=False)["Total"].mean().idxmax()


# 5️⃣ List students scoring above 85 in all subjects

lst_stds = df[df[["Math","English","Science","History"]]>85].all(axis=1)
print(lst_stds)
```
